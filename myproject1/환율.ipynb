```python
import matplotlib.pyplot as plt

# 국가별 데이터 선택
korea1 = kmt_df.loc['대한민국']
korea2 = kmt_df1.loc['대한민국(Won)']

# subplots로 두 개의 그래프 영역을 만듦
fig, ax = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

# 첫 번째 subplot에 그래프 그리기
ax[0].plot(korea1.index, korea1.values, color='blue')
ax[0].set_title('대한민국 외환 보유액')
ax[0].set_ylabel('달러')

# 두 번째 subplot에 그래프 그리기
ax[1].plot(korea2.index, korea2.values, color='red')
ax[1].set_title('대한민국 환율')
ax[1].set_ylabel('원(won)')

# 전체 그래프 제목과 축 이름 설정
plt.xlabel('연도')
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.95, wspace=0.4, hspace=0.4)
# 그래프 출력
plt.show()
```
